```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#DISTRICT SUMMARY
#Calculate the total number of schools
total_schools = len(school_data_complete["school_name"].unique())

#Calculate the total number of students
total_students = (school_data_complete["student_name"].count())

#Calculate the total budget
total_budget = sum(school_data_complete["budget"].unique())

#Calculate the average math score
average_math = school_data_complete["math_score"].mean()

#Calculate the average reading score
average_reading = school_data_complete["reading_score"].mean()

#Calculate the percentage of students with a passing math score (70 or greater)
passing_math = (school_data_complete[school_data_complete["math_score"]>=70]["student_name"].count()/total_students)*100

#Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading = (school_data_complete[school_data_complete["reading_score"]>=70]["student_name"].count()/total_students)*100

#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_passing = (school_data_complete[ 
       (school_data_complete['reading_score'] >= 70) & 
       (school_data_complete['math_score'] >= 70) ]["student_name"].count() / total_students ) * 100
```

```python
#Create a dataframe to hold the above results
district_summary = pd.DataFrame({"Total Schools": [total_schools],
                                 "Total Students": [total_students],
                                 "Total Budget": [total_budget],
                                 "Average Math Score": [average_math],
                                 "Average Reading Score": [average_reading],
                                 "% Passing Math": [passing_math],
                                 "% Passing Reading": [passing_reading],
                                 "% Overall Passing Score": [overall_passing]})

#clean data
district_summary["Total Students"] = district_summary["Total Students"].map("{:,.0f}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:.2f}%".format)
district_summary["% Overall Passing Score"] = district_summary["% Overall Passing Score"].map("{:.2f}%".format)

#display DISTRICT SUMMARY data frame
district_summary
```

```python
#SCHOOL SUMMARY

#group data by school name
group_schools = school_data_complete.groupby(['school_name'])

#calculate total students for each school
students_school = group_schools.size()

#calculate different school types
school_types = group_schools['type'].first()

#calculate total budget
total_school_budget = group_schools['budget'].first()

#calculate per student budget
budget_student = total_school_budget/students_school

#calculate average math scores
math_average = group_schools['math_score'].mean()

#calculate average reading score
reading_average = group_schools['reading_score'].mean()

#calculate % passing math
passing_math_total = school_data_complete[school_data_complete['math_score']>=70].groupby(['school_name']).size()
passing_math_percent = (passing_math_total/students_school)*100

#calculate % passing reading 
passing_reading_total = school_data_complete[school_data_complete['reading_score']>=70].groupby(['school_name']).size()
passing_reading_percent = (passing_reading_total/students_school)*100

#calculate % overall passing
overall_passing_percent = school_data_complete[ 
       (school_data_complete['reading_score'] >= 70) & 
       (school_data_complete['math_score'] >= 70) ].groupby(['school_name']).size()
overall_passing_percent = (overall_passing_percent/students_school)*100
overall_passing_percent
```

```python
#Create a dataframe to hold the above results
school_summary = pd.DataFrame({"School Type": school_types,
                               "Total Students": students_school,
                               "Total School Budget": total_school_budget,
                               "Per Student Budget": budget_student,
                               "Average Math Score": math_average,
                               "Average Reading Score": reading_average,
                               "% Passing Math": passing_math_percent,
                               "% Passing Reading": passing_reading_percent,
                               "% Overall Passing": overall_passing_percent})

#clean data
school_summary["Total Students"] = school_summary["Total Students"].map("{:,.0f}".format)
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,.2f}".format)
school_summary["Average Math Score"] = school_summary["Average Math Score"].map("{:.2f}".format)
school_summary["Average Reading Score"] = school_summary["Average Reading Score"].map("{:.2f}".format)
school_summary["% Passing Math"] = school_summary["% Passing Math"].map("{:.2f}%".format)
school_summary["% Passing Reading"] = school_summary["% Passing Reading"].map("{:.2f}%".format)
school_summary["% Overall Passing"] = school_summary["% Overall Passing"].map("{:.2f}%".format)

school_summary
```

```python
#Top Performing Schools (By % Overall Passing)

#Sort and display the top five performing schools by % overall passing.
school_summary = school_summary.sort_values("% Overall Passing", ascending=False)
school_summary.head()
```

```python
#Bottom Performing Schools (By % Overall Passing)

#Sort and display the five worst-performing schools by % overall passing.
school_summary = school_summary.sort_values("% Overall Passing", ascending=True)
school_summary.head()
```

```python
#Math Scores by Grade

#Create a pandas series for each grade. Hint: use a conditional statement.
#Group each series by school
math_9th = school_data_complete.loc[school_data_complete["grade"] == "9th"].groupby(["school_name"])['math_score'].mean()
math_10th = school_data_complete.loc[school_data_complete["grade"] == "10th"].groupby(["school_name"])['math_score'].mean()
math_11th = school_data_complete.loc[school_data_complete["grade"] == "11th"].groupby(["school_name"])['math_score'].mean()
math_12th = school_data_complete.loc[school_data_complete["grade"] == "12th"].groupby(["school_name"])['math_score'].mean()

#Combine the series into a dataframe
math_grades = pd.DataFrame({"9th": math_9th, 
                            "10th": math_10th, 
                            "11th": math_11th, 
                            "12th": math_12th })

#clean data
math_grades["9th"] = math_grades["9th"].map("{:.2f}".format)
math_grades["10th"] = math_grades["10th"].map("{:.2f}".format)
math_grades["11th"] = math_grades["11th"].map("{:.2f}".format)
math_grades["12th"] = math_grades["12th"].map("{:.2f}".format)

math_grades = math_grades.rename_axis([""])

math_grades
```

```python
#Reading Scores by Grade

#Create a pandas series for each grade. Hint: use a conditional statement.
#Group each series by school
reading_9th = school_data_complete.loc[school_data_complete["grade"] == "9th"].groupby(["school_name"])['reading_score'].mean()
reading_10th = school_data_complete.loc[school_data_complete["grade"] == "10th"].groupby(["school_name"])['reading_score'].mean()
reading_11th = school_data_complete.loc[school_data_complete["grade"] == "11th"].groupby(["school_name"])['reading_score'].mean()
reading_12th = school_data_complete.loc[school_data_complete["grade"] == "12th"].groupby(["school_name"])['reading_score'].mean()

#Combine the series into a dataframe
reading_grades = pd.DataFrame({"9th": reading_9th, 
                               "10th": reading_10th, 
                               "11th": reading_11th, 
                               "12th": reading_12th })

#clean data
reading_grades["9th"] = reading_grades["9th"].map("{:.2f}".format)
reading_grades["10th"] = reading_grades["10th"].map("{:.2f}".format)
reading_grades["11th"] = reading_grades["11th"].map("{:.2f}".format)
reading_grades["12th"] = reading_grades["12th"].map("{:.2f}".format)

reading_grades = reading_grades.rename_axis([""])

reading_grades
```

```python
#Scores by School Spending

#create bins
bins = [0, 584, 629, 644, 675]
group_names = ["<$584", "$585-629", "$630-644", "$645-675"]

#use pd.cut to put new bins into summary data frame
school_summary["Spending Ranges (Per Student)"] = pd.cut(budget_student, bins, labels=group_names)

school_summary

#Average Reading Score
#% Passing Math
#% Passing Reading
#Overall Passing Rate (Average of the above two)
```

```python
school_summary['Average Math Score'] = school_summary['Average Math Score'].astype(float)
school_summary['Average Reading Score'] = school_summary['Average Reading Score'].astype(float)
school_summary['% Passing Math'] = school_summary['% Passing Math'].str[0:-1].astype(float)
school_summary['% Passing Reading'] = school_summary['% Passing Reading'].str[0:-1].astype(float)
school_summary['% Overall Passing'] = school_summary['% Overall Passing'].str[0:-1].astype(float)

school_summary.groupby('Spending Ranges (Per Student)')[['Average Math Score',
                                                         'Average Reading Score', '% Passing Math', 
                                                         '% Passing Reading', '% Overall Passing', 'Spending Ranges (Per Student)']].mean()
```

```python
#Scores by School Size
#create bins
bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

#use pd.cut to put new bins into summary data frame
school_summary["School Size"] = pd.cut(students_school, bins, labels=group_names)

school_summary.groupby('School Size')[['Average Math Score',
                                       'Average Reading Score', 
                                       '% Passing Math', 
                                       '% Passing Reading', 
                                       '% Overall Passing', 
                                       'School Size']].mean()
```

```python
#Scores by School Type

school_summary.groupby('School Type')[['Average Math Score',
                                       'Average Reading Score', 
                                       '% Passing Math', 
                                       '% Passing Reading', 
                                       '% Overall Passing', 
                                       'School Type']].mean()
```

DATA ANALYSIS

Overall, Charter schools performed better than District schools. 90 percent of students at charter schools passed both exams while only 53 percent of students from district schools passed both. 

The per student budget for the lowest performing schools is generally higher than the per student budget for the highest schools, which means the budget made little difference in determining the success of a student. 

The more significant factor seems to be the school size, smaller schools had better passing rates than schools with more students. This likely suggests that smaller class size and more availability from their teachers proved to be more helpful to them passing the exams.

Each grade performed pretty consistently at each individual school.
